# Two-Way ANOVA (Factorial Design)

```python
# Import required libraries
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

An automotive research institute wants to determine whether fuel efficiency (in miles per gallon, MPG) is significantly affected by two factors:

- *Vehicle Class*: Sedan vs. SUV

- *Engine Type*: Hybrid vs. Conventional

They collected MPG data for vehicles representing all combinations of these two factors. The data are summarized below:

```python

# Create dataset
data = pd.DataFrame({
    'Model': ['A', 'A', 'B', 'B', 'C', 'C', 'D', 'D', 'E', 'E', 'F', 'F'],
    'Class': ['Sedan', 'Sedan', 'Sedan', 'Sedan', 'SUV', 'SUV', 'SUV', 'SUV', 'Sedan', 'Sedan', 'SUV', 'SUV'],
    'EngineType': ['Hybrid', 'Conventional', 'Hybrid', 'Conventional',
                   'Hybrid', 'Conventional', 'Hybrid', 'Conventional',
                   'Hybrid', 'Conventional', 'Hybrid', 'Conventional'],
    'MPG': [42, 35, 44, 34, 32, 26, 34, 25, 43, 36, 33, 27]
})
print(data)
```

At the
$\alpha = 0.05$ significance level, test for significant effects of:

- Vehicle Class

- Engine Type

- Interaction between Class and Engine Type

```python
# Fit two-way ANOVA model with interaction
model = ols('MPG ~ C(Class) + C(EngineType) + C(Class):C(EngineType)', data=data).fit()

# Two-way ANOVA table
anova_table = sm.stats.anova_lm(model, typ=2)

# Display ANOVA results
print(anova_table.round(4))
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
tukey_results = pairwise_tukeyhsd(endog=data["MPG"],
                                  groups=data["Class"] + "_" + data["EngineType"])
print(tukey_results.summary())
```
