You’re an operations analyst for AT&T.  Long-distance telephone calls are normally distributed with $\mu$ = 8 min. and $\sigma$  = 2 min.  If you select a random sample of 25 calls, what percentage of the calls would be between 7.8 & 8.2 minutes?


First, we sample the values of the call duration from normal distribution with the required mean and standard deviation:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu = 8
sigma = 2

# Generate 25 normally distributed random variables with mean 8 and standard deviation 2 and print it out
durations = np.random.normal(loc=mu, scale=sigma, size=25)
print("Call durations:", durations)

#Create and print a true-false vector 
bool_vector = (durations > 7.8) & (durations <= 8.2)
print("True-False vector for calls between 7.8 and 8.2 minutes:", bool_vector)

# Find the proportion of calls between 7.8 and 8.2 minutes
proportion = np.mean(bool_vector)

# Print the proportion
print("Proportion of calls between 7.8 and 8.2 minutes: ", proportion)

```

The commands below will give you a prettier vector output:

```python
# Print the array of generated call durations, values separated by commas, no brackets
print("Call durations:", ', '.join(map(str, durations)))

# Print the True-False vector, values separated by commas, no brackets
print("True-False vector for calls between 7.8 and 8.2 minutes:", ', '.join(map(str, bool_vector)))
```

We can also create a histogram of the call durations:

```python
plt.hist(durations, bins=10, color='purple', alpha = 0.2, edgecolor='black')

# Add labels and title
plt.xlabel('Call Duration (minutes)')
plt.ylabel('# of calls')
plt.title('Histogram of Call Durations')

# Show the plot
plt.show()
```

Since the distribution is normal, the theoretical proportion of the calls in that range is $P(X \leq 8.2) - P(X \leq 7.8)$:

```python
prob = norm.cdf(8.2, loc=mu, scale=sigma) - norm.cdf(7.8, loc=mu, scale=sigma)
print("Theoretical probability of call duration between 7.8 and 8.2 minutes: ", prob)
```
