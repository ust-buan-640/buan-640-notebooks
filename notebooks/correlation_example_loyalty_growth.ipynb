```python
import numpy as np
import pandas as pd
import scipy.stats as stats


#bring the file noodles loyalty card into a python dataframe
df = pd.read_csv('https://raw.githubusercontent.com/ust-buan-640/buan-640-notebooks/main/data/noodles-loyalty-card.csv')

#print out the first few rows and columns of data to look at the output to ensure it is ok to use
print(df.head())
```

```python
#display the scatterplot


ax1 = df.plot.scatter(x='Loyalty_pct',
                      y='Growth_pct',
                      c='DarkBlue')
```

```python
#calculate pearson correlation coefficients for growth% and loyalty%

coeff = stats.pearsonr(df['Loyalty_pct'], df['Growth_pct'])
print(coeff)
```
