You’re a marketing analyst for retail store chain.  The chain had teddy bears on sale last week.  The weekly sales of bears sold in 10 stores that you sampled was: 8  11  0   4   7   8   10   5   8   3. The historical data shows that the distribution of sales across the stores is approximately normal. At the .05 level of significance, is there enough evidence to conclude that the average bear sales per store is more than 5?


```python
import scipy.stats as stats

sales = [8, 11, 0, 4, 7, 8, 10, 5, 8, 3]

# Define the hypothesized population mean (null hypothesis) and significance level
population_mean = 5
alpha = 0.05

# Perform a one-sample t-test
t_statistic, p_value = stats.ttest_1samp(sales, population_mean, alternative = 'two-sided' )

if p_value < alpha:
    print("Reject the null hypothesis: There is enough evidence to conclude that the average bear sales per store is different than 5.")
else:
    print("Fail to reject the null hypothesis: There is not enough evidence to conclude that the average bear sales per store is different than 5.")

# Print t-statistic and right-tailed p-value for reference
print(f"t-statistic: {t_statistic}")
print(f"Two-tailed test p-value: {p_value}")
```
