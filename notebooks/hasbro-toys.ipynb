```python
import matplotlib.pyplot as plt
import scipy.stats as stats
import pandas as pd
from statsmodels.formula.api import ols 
```

```python
ad = [1, 2, 3, 4, 5]  # Ad expenditure in 100$
sales = [1, 1, 2, 2, 4]  # Sales in units

# Create DataFrame
df = pd.DataFrame({'ad': ad, 'sales': sales})

# Fit linear model
lm = ols('sales ~ ad', data=df).fit()

```

```python
plt.figure(figsize=(6, 3))
plt.scatter(ad, sales, color = 'purple', s = 200, label='Data points')

plt.xlim(0)  # Ensure x-axis starts at 0
plt.ylim(0)  # Ensure y-axis starts at 0

# Add labels and title
plt.xlabel('Ad Expenditure (100$)')
plt.ylabel('Sales (Units)')

#Save the figure
plt.savefig('img-output/hasbro.png', bbox_inches='tight')
```

```python

# Print the regression coefficients
print(lm.summary())
```

```python
r, p_value = stats.pearsonr(ad, sales)
print("Correlation:", r)
print("P-value:", p_value)
```
